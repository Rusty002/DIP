## Storage size of an image: 

$size (bits) = rows \times columns \times colors(rgb) \times grey levels$

$size (bits) = 256 \times 384 \times 3 \times 8 $

```python
import numpy as np
#size of an image
img_shape = img.shape
print("Image resolution :",img_shape)
img_size = np.prod(img_shape)
print("Storage size in bits: ", img_size*8)

print("Storage size in bytes: ", img_size)
```

# Rescale image : 
## Rescale operation resizes an image by a given scaling factor. The scaling factor can either be a single floating point value, or multiple values - one along each axis.

```python
from skimage import data, color
from skimage.transform import rescale, resize, downscale_local_mean
image = color.rgb2gray(data.astronaut())
image_rescaled = rescale(image, 1.0 / 4.0)

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original image')

plt.subplot(1, 2, 2)
image_rescaled = rescale(image, 1.0 / 4.0)
plt.imshow(image_rescaled, cmap='gray' )
plt.title('rescaled 1/4')
```

# Resize Image
## Resize serves the same purpose, but allows to specify an output image shape instead of a scaling factor.

```python
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original image')
print(image.shape)
image_resized = resize(image, (image.shape[0] / 4, image.shape[1] / 8))
plt.subplot(1, 2, 2)
image_rescaled = rescale(image, 1.0 / 4.0)
plt.imshow(image_resized, cmap='gray' )
plt.title('resized 1/4')
```

## 1-D Interpolation

```python
from scipy.interpolate import interp1d
x = np.linspace(0, 10, num=11, endpoint=True)
y = [5,3,7,10,5,7,10,10,7,8,9]
f = interp1d(x, y)
f2 = interp1d(x, y, kind='quadratic')
f3 = interp1d(x,y, kind='cubic')
xnew = np.linspace(0, 10, num=41, endpoint=True)
plt.plot(x, y, 'o', xnew, f(xnew), '-', xnew, f2(xnew), '--', xnew, f3(xnew), '--')
plt.legend(['data', 'linear', 'quadratic', 'cubic'], loc='best')
plt.show()
```

# 2D Interpolation

```python
from scipy.interpolate import interp2d
x = np.linspace(0, 4, 13)
y = np.array([0, 2, 3, 3.5, 3.75, 3.875, 3.9375, 4])
X, Y = np.meshgrid(x, y)
Z = np.sin(np.pi*X/2) * np.exp(Y/2)

x2 = np.linspace(0, 4, 65) 
y2 = np.linspace(0, 4, 65)
f = interp2d(x, y, Z, kind='cubic')
Z2 = f(x2, y2)

fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].pcolormesh(X, Y, Z, cmap='jet')

X2, Y2 = np.meshgrid(x2, y2)
ax[1].pcolormesh(X2, Y2, Z2, cmap='jet')
plt.show()
```
